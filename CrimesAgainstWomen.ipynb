```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasheets into dataframes
df1 = pd.read_csv('/CrimesOnWomenData.csv')
df2 = pd.read_csv('/Indias GDP (Statewise).csv')

# Merge or align the dataframes based on a common column (e.g., 'ID')
merged_df = pd.merge(df1, df2, on='State')
merged_df.head()
```

image.png

```python
numerical_df = merged_df[['Rape', 'K&A', 'DD', 'AoW','AoM', 'DV','WT','Nominal GDP(Trillion INR)']] # Select numerical columns
# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar = True)
plt.title('Correlation Heatmap')
plt.show()
```

It can be analysed that women trafficking has some 0.6 correlation with nominal GDP

We will just plot the graphs against each of the to show if they are related.

```python
crime_columns = ['Rape', 'K&A', 'AoW','AoM', 'DD','DV', 'WT']  # Update this list with actual crime columns from your dataset
gdp_column = 'Nominal GDP(Trillion INR)'  # Assuming this is the column name for GDP

plt.figure(figsize=(12, 8))

for i, crime in enumerate(crime_columns):
    plt.subplot(2, 4, i + 1)  # Adjust subplot configuration based on the number of plots
    plt.scatter(merged_df[gdp_column], merged_df[crime], alpha=0.8)
    plt.ylabel(crime)
    plt.xlabel('GDP')
    plt.title(f'scatter plot of {crime} vs GDP')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Except for rape, assault of women and women trafficking which had higher numbers even with high GDP of states most other crimes have diminished number with high nominal GDP. Quite definitely dowry deaths were less in higher GDP states. Also observed that the scatter plots were denser between 0-20 GDP .
